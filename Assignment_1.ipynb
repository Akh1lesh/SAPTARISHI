#Import required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IPL_Dataset.csv')
```

```python
df
```

# Q1. What is the maximum number of matches played by an individual player in a season?

```python
df[df['Mat'] == df['Mat'].max()]
```

#Q2. What is the highest average of a player in a season?

```python
df[df['Avg'] == df['Avg'].max()]
```

#Q3. What is the average number of matches played by the players in a season?

```python
df['Mat'].mean()
```

#Q4. Please check the correlation between all the features.

```python
correlation = df.corr()
labels = np.array(df.columns)
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation,annot = True,xticklabels = labels, yticklabels = labels)
plt.title("Correlation matrix plot")
```

# Q5. Check the list of players who has an average greater than 50 as well strike rate above 120.

```python
df[(df['Avg']>50) & (df['SR'] > 120)]
```

# Q6. Please check the list of players who has an average greater than 40 and balls faced above 100.

```python
df[(df['Avg']>40) & (df['BF'] > 100)]
```

# Q7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.

```python
df[(df['Runs'] > 300) & (df['Avg'] > 45) & (df['SR'] > 130)]
```

# Q8. Players who scored alteast one century in this season.

```python
df[df['100'] > 0]
```

# Q9. Players who scored atleast 4 half centuries in this season.

```python
df[df['50']>=4]
```

# Q10. Check the list of players who hit more than 45 boundARIES and more than 10 sixes in this season.

```python
df[(df['4s'] + df['6s'] > 45) & (df['6s'] > 10)]
```

# Q11. Plot a histogram of number of matches played in a season by players.

```python
plt.figure(figsize=(10,10))
plt.hist(df['Mat'],bins = 100)
plt.xlabel('Number of Matches')
plt.ylabel('count')
plt.title('Number of matches played in a season by Players')
```

# Q12. Plot the histogram of balls faced by players.

```python
plt.figure(figsize = (10,10))
plt.hist(df['BF'],bins = 100)
plt.xlabel('Balls Faced')
plt.ylabel('Count')
plt.title("Histogram of Balls Faced by players")
```

# Q13. Top 20 players with most runs in a season.

```python
df.sort_values(by = ['Runs'],ascending = False).head(20)
```

# Q14. Bar chart for top 20 players with best strike rate in a season.

```python
df1 = df.sort_values(by = ['SR'],ascending = False).head(20)
```

```python
df1_SR = df1['SR']
df1_Player = df1['Player']
```

```python
plt.figure(figsize = (10,10))
plt.bar(df1_Player,df1_SR)
plt.xticks(rotation = 'vertical')
plt.xlabel('Player Name')
plt.ylabel('Strike Rate')
plt.title("Top 20 players with Best Strike Rate")
```

# Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
df_Runs = df['Runs']
df_BF = df['BF']
print(df_Runs.max())
print(df_BF.max())
```

```python
plt.figure(figsize = (10,10))
plt.scatter(df_BF,df_Runs)
plt.xlabel('Balls faced by Players')
plt.ylabel('Runs scored by Players')
plt.title("Scatter plot of Runs scored by Player V/s Balls faced by Players")
```

# Q17. KDE plot for averages on a season.

```python
plt.figure(figsize = (10,10))
sns.kdeplot(df['Avg'],fill = True)
plt.title("KDE Plot for averages on a season")
```
